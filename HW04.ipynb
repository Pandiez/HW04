import numpy as np


def bootstrap_ci(data, stat_func=np.mean, n_bootstrap=10000, ci=95):
    """
    data: array-like, the original sample data
    stat_func: function, the statistic to compute (default is np.mean)
    n_bootstrap: int, the number of bootstrap samples (default is 10,000)
    ci: float, the confidence interval level (default is 95%)
    """
   
    bootstrap_stats = []
    for _ in range(n_bootstrap):
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        bootstrap_stat = stat_func(bootstrap_sample)
        bootstrap_stats.append(bootstrap_stat)
    
  
    bootstrap_stats_sorted = np.sort(bootstrap_stats)
    
 
    lower_bound = np.percentile(bootstrap_stats_sorted, (100 - ci) / 2)
    upper_bound = np.percentile(bootstrap_stats_sorted, 100 - (100 - ci) / 2)
    
    return lower_bound, upper_bound


data_sample = np.random.randn(100) 


ci_mean = bootstrap_ci(data_sample, stat_func=np.mean)
print("95% Bootstrap CI for the mean:", ci_mean)


ci_median = bootstrap_ci(data_sample, stat_func=np.median)
print("95% Bootstrap CI for the median:", ci_median)


ci_variance = bootstrap_ci(data_sample, stat_func=np.var)
print("95% Bootstrap CI for the variance:", ci_variance)

When creating confidence intervals, we must distinguish between the population parameter and the sample statistic because they stand for distinct things:

Population parameter: This is the real value that we are interested in knowing (for example, the true population mean or median), but we are typically unable to measure it directly due to the lack of data from the complete population.

Sample statistic: This is the result of computing the sample mean or sample median using the sample data. Although it fluctuates from sample to sample, it is our best approximation of the population parameter.

Why this matters for confidence intervals: 
Based on the sample statistic, a confidence interval gives us an estimate of the range that includes the true population parameter.
The confidence interval illustrates the degree of uncertainty surrounding the sample statistic, which is merely an estimate of the population parameter. It's critical to understand that the interval describes more than just the sample data; it provides a likely range for the unknown population value.

1. What is the bootstrapping process? Bootstrapping is a resampling technique that involves regularly taking replacement samples from a given dataset in order to estimate the distribution of a statistic. The actions are as follows:
Resample the information: Take replacement samples at random from the original dataset, making sure each resampled set has the same dimensions as the original sample. Determine the statistic by: Determine the relevant statistic (such as the mean or median) for every resampled group. Continue as before: To generate a distribution of the statistic, repeat this resampling and computing procedure a number of times (e.g., 1,000 or 10,000 times). Examine the outcomes: To estimate confidence intervals or evaluate the statistic's variability, use the distribution of the resampled statistics.

2. What is bootstrapping's primary goal? Bootstrapping is primarily used to estimate the variability or uncertainty of a sample statistic (such as the variance, mean, or median) in situations where the underlying population distribution is unknown or in which it is difficult to justify typical parametric assumptions (such as normality).

If zero falls inside the confidence interval, even though the sample mean is not zero, it suggests that the data is not strong enough to completely rule out the possibility that the genuine effect is zero. Due to our inability to declare with confidence that the medicine has a non-zero impact based on the sample, we are left "failing to reject the null hypothesis".

Contrary Conclusion (Rejecting the Null Hypothesis): If 0 is not included in the confidence interval, we would reject the null hypothesis.
